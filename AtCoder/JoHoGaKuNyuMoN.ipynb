```python
x = int(input("点数を入力:"))
if x > 60 and x <= 80: # ここに「=」を追加し、80点を処理できるようにした
    print("B")
elif x > 80:
    print("A")
else:
    print("C")
```

```python
x = int(input("整数を入力:"))
for n in range(1, x + 1): # 「x + 1」に変更し、その入力した値も表示されるようにした
     if n % 2 != 0:
         print(n)
```

```python
# 1. 数学の点数 X を入力
try:
    x = int(input("数学の点数を入力してください (X): "))
except ValueError:
    print("無効な入力です。整数を入力してください。")
    exit()

# 2. 英語の点数 Y を入力
try:
    y = int(input("英語の点数を入力してください (Y): "))
except ValueError:
    print("無効な入力です。整数を入力してください。")
    exit()

# 3. 条件分岐による判定
if x > y:
    difference = x - y
    print(f" {difference} 点")
elif y > x:
    print("数学強化")
else:
    print("許容")
```
